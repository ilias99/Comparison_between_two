```python
import ntpath
import re

# function that extract name file without extension from path
def path_leaf(path):
    head, tail = ntpath.split(path)
    return tail or ntpath.basename(head)

# find text between 'first' substring and 'last' substring given a string s
def find_between( s, first, last ): 
    try:
        start = s.index( first ) + len( first )
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return "" 

from termcolor import colored

def removekey(d, key):
    r = dict(d)
    del r[key]
    return r
```

```python
# function that read .txt files and filtering what is not necessary like http links, headers, "copyright" etc
# return dictionary in the for:  
# dict[name_file] = text of name_file as string

def clean_text_files(allfiles,name_var):
    
    a = {}
    b = {}
    
    i = 0
    for file in allfiles:
    
        lines = [] #Declare an empty list named "lines"
        with open (file, 'rt', encoding="ANSI") as in_file:    #Open file in allfiles for reading of text data.
            
            for line in in_file: #For each line of text store in a string variable named "line", and
                lines.append(line)  #add that line to our list of lines.

        a[name_vars[i]] = ''
        
        for line in lines:

            if path_leaf(file)[:4] and 'SRM' and '2018' not in line: #remove first line of the pdf irrelevant
                if '/2018' not in line:
                    if '© ' not in line:
                        
                        a[name_vars[i]] = a[name_vars[i]] + line
                    
        
        a[name_vars[i]] = re.sub(r'^https?:\/\/.*[\r\n]*', '', a[name_vars[i]], flags=re.MULTILINE) #remove http links
        a[name_vars[i]] = re.sub(r'^https?:\/\/.*[\r\n]*', '', a[name_vars[i]], flags=re.MULTILINE) #remove http links
        
        #b[name_vars[i]] = a[name_vars[i]] 
        a[name_vars[i]] = a[name_vars[i]][:a[name_vars[i]].index( "Figure 001 / " )] #extract relevant text till caption of figures 
        
        i = i+1
    
    return a
```

```python
def extract_paragraphs(text):

	#titles of paragraphs
	#update with new cases if you find new ones
	wv = ['C. Weight Variant Applicability','C. WV Applicability','Weight Variant Applicability']
	lr = ['D. Limits/Restrictions','D. Limits and Restrictions']
	jsi = ['G. Job Set-Up Information', '(1) Material Reference:']
	ad = ['A. Allowable Dents on Fuselage Skin Plates','A. Allowable Dents','A. Allowable dent limits and dent dress-out','A. FR24 thru FR35 - Allowable Dents','A. Allowable Damage','Allowable dent limits and dent dress-out\n(1)']
	warn_caut = ['WARNING:','CAUTION:']
	caut = 'CAUTION:'
	gen = ['A. General'] 
	aoa = ['B. Area of Applicability']
	rc = ['E. Repair Category']
	ins = ['F. Inspections']
	
	

	paragraph_ls = {} 

	paragraph_ls['Caution'] = (find_between(text,find_title(warn_caut,text),"1. Task Supporting Data").replace(find_title(warn_caut,text),'').replace(caut,'')).lower()
	paragraph_ls['General'] = (find_between(text,find_title(gen,text),find_title(aoa,text))).lower()
	paragraph_ls['Area of Applicability'] = (find_between(text,find_title(aoa,text),find_title(wv,text))).lower()
	paragraph_ls['Weight Variant Applicability'] = (find_between(text,find_title(wv,text),find_title(lr,text))).lower()
	paragraph_ls['Limits/Restrictions'] = (find_between(text,find_title(lr,text),find_title(rc,text))).lower()
	paragraph_ls['Repair Category'] = (find_between(text, find_title(rc,text), find_title(ins,text))).lower()
	paragraph_ls['Inspections'] = (find_between(text,find_title(ins,text),find_title(jsi,text))).lower()
	paragraph_ls['Job Set-Up Information'] = (find_between(text,find_title(jsi,text),find_title(ad,text))).lower()
	paragraph_ls['Allowable D'] = (text[text.index(find_title(ad,text)) + len(find_title(ad,text)):]).lower()
	
	return paragraph_ls
```

```python
#given two dict in input
#dict1[title paragraph 1] = text paragraph 1
#dict2[title paragraph 2] = text paragraph 2
#give in output dict3[text paragraph 1] = text paragraph 2

def proc(d1,d2):

    output = {} # a strange dictionary: output[text of paragraph of first doc] = text of paragraph of second text

    title_paragraph = [] #titles of paragraphs

    for key in d1.keys() & d2.keys():
    
        output.setdefault(d1[key], []).append(d2[key])
        title_paragraph.append(key)
        
    for key,value in output.items():
        output[key] = (''.join(value)).replace('\n',' ').replace('. ','\n')
        
    
    return output, title_paragraph 
```

```python
#split text in pairs of word, return list of pairs of words in the same order they appear in text 

def split_string_pairs(text):
    
    
    #text = text.replace(' the ',' ').replace('. ',' ').replace('(',' ').replace(')',' ').replace(', ',' ').replace("'","").replace("'","").replace(':','')
    text = text.replace(',',' ')
    
    words = text.split() #note that split without arguments splits on whitespace
    pairs = [words[i]+' '+words[i+1] for i in range(len(words)-1)]
    
    new_pairs=[]
    
    for index, element in enumerate(pairs):
        
        if '. ' not in element:
            new_pairs.append(element)
      
    for index,element in enumerate(new_pairs):
        
        if '.' in element:
            new_pairs[index] = new_pairs[index].replace('.','')
            
    return new_pairs

#given two lists of pairs of words return list of common pairs  
def compare(list1,list2):
    
    common_pairs=[]
    for element in list1:
        for element2 in list2:
            if element == element2:
                if element not in common_pairs: 
                    common_pairs.append(element)
                    
    return common_pairs

def clean_line(line): # filtering some of punctuation not needed
    
    line = line.replace(' the ',' ').replace('(',' ').replace(')',' ').replace(', ',' ').replace("'","").replace("'","").replace(':','').replace(' "',' " ').replace('" ',' " ').replace('"',' "')
    line = re.sub(' +',' ', line)
    
    return line 

def clean_par(par):
    return par.replace(' the ',' ').replace('. ',' ').replace('(',' ').replace(')',' ').replace(', ',' ').replace("'","").replace("'","").replace(':','').replace(' "',' " ').replace('" ',' " ').replace('"',' "')
```
