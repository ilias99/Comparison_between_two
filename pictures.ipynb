```python
def split_string_pairs(text):
    
    
    #text = text.replace(' the ',' ').replace('. ',' ').replace('(',' ').replace(')',' ').replace(', ',' ').replace("'","").replace("'","").replace(':','')
    text=text.replace(',',' ')
    
    words = text.split() #note that split without arguments splits on whitespace
    pairs = [words[i]+' '+words[i+1] for i in range(len(words)-1)]
    
    new_pairs=[]
    
    for index, element in enumerate(pairs):
        
        
        #if '- ' not in element:
         #   if ' -' not in element:
        if '. ' not in element:
            new_pairs.append(element)
    
    #print(new_pairs)
    #print(pairs)
      
    for index,element in enumerate(new_pairs):
        
        if '.' in element:
            new_pairs[index] = new_pairs[index].replace('.','')
    
   # n = []
    
    #for element in new_pairs:
     #   if 'the' not in element:
            
      #      n.append(element)          
    
    return new_pairs

def compare(list1,list2):
    
    common_pairs=[]
    for element in list1:
        for element2 in list2:
            if element == element2:
                if element not in common_pairs: 
                    common_pairs.append(element)
                    
    return common_pairs

def clean_line(line):
    
    line = line.replace(' the ',' ').replace('(',' ').replace(')',' ').replace(', ',' ').replace("'","").replace("'","").replace(':','').replace(' "',' " ').replace('" ',' " ').replace('"',' "')
    line = re.sub(' +',' ', line)
    
    return line 

def clean_par(par):
    return par.replace(' the ',' ').replace('. ',' ').replace('(',' ').replace(')',' ').replace(', ',' ').replace("'","").replace("'","").replace(':','').replace(' "',' " ').replace('" ',' " ').replace('"',' "')
```

```python
def count_common_word(sentence1,sentence2):
    
    n = 0
    
    words1 = set(sentence1.split())
    for word in set(sentence2.split()):
    # strip some chars here, e.g. as in [1]
        if word in words1:
            n += 1
            
    return n 
    
  
```
